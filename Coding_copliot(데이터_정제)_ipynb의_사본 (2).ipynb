## 데이터 전처리

```python
import json

with open('/content/data_Inspection - 시트12 (1).csv', 'r', encoding='utf-8') as file:
  data = json.load(file)

lengths = []

# 각 행에 대해 반복하면서 글자 수를 계산합니다.
for row in data:
    # JSON 객체(행)를 문자열로 변환하고 길이를 측정합니다.
    row_length = len(json.dumps(row, ensure_ascii=False))
    lengths.append(row_length)

# 결과를 출력합니다.
longest_length = max(lengths)
index_of_longest = lengths.index(longest_length)
data[index_of_longest]
#longest_length
```

```python
import json

with open('/content/final-data (3).json', 'r', encoding='utf-8') as file:
    data = json.load(file)

byte_lengths = []

# 각 행에 대해 반복하면서 바이트 길이를 계산합니다.
for row in data:
    # JSON 객체(행)를 문자열로 변환하고, 이를 utf-8로 인코딩하여 바이트로 변환합니다.
    row_bytes = json.dumps(row, ensure_ascii=False).encode('utf-8')
    # 변환된 바이트의 길이를 계산합니다.
    row_byte_length = len(row_bytes)
    byte_lengths.append(row_byte_length)

# 결과를 출력합니다.
longest_byte_length = max(byte_lengths)
index_of_longest = byte_lengths.index(longest_byte_length)
data[index_of_longest]
# longest_byte_length
```

```python
from transformers import AutoTokenizer

# 모델 ID 지정
model_id = "TinyPixel/CodeLlama-7B-Python-bf16-sharded"
# 해당 모델에 대한 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained(model_id)
text = '{text: Below is an instruction that describes a task. Write a response that appropriately completes the request. ### Instruction: 리스트의 생성 방법을 자세하게 설명해 주실 수 있나요? ### Response: 리스트에 관련해서 질문해 주셨네요. 리스트는 파이썬에서 여러 항목을 하나의 변수에 저장하는 자료형으로, 순서가 있는 데이터를 다룰 때 사용됩니다. 리스트를 생성하려면 대괄호\'[ ]\' 안에 항목들을 쉼표로 구분하여 나열하면 됩니다. 예를 들어, "my_list = [1, 2, 3, 4, 5]"와 같이 리스트를 정의할 수 있으며, 이렇게 생성된 리스트는 각 항목에 인덱스를 사용하여 접근하고 수정할 수 있습니다. 리스트는 다양한 데이터를 담을 수 있어 데이터의 저장과 관리에 유용하며, 프로그래밍에서 자주 활용됩니다.}'

tokens = tokenizer.tokenize(text)
number_of_tokens = len(tokens)

number_of_tokens
```

```python
# import pandas as pd
import json

# # CSV 파일을 다시 읽어들임
# file_path = '/content/24answers.csv'
# df = pd.read_csv(file_path, sep='\t', encoding='cp949')

# # 이 CSV 파일을 위의 코드에 맞는 JSON 구조로 변환
# # 각 행을 'qas' 키 안에 'question'과 'answers'로 넣음
# print(df.columns)

# 먼저 모든 행을 'qas' 형식으로 변환
qas_list = df.apply(lambda row: {'qas': [{'question': row['Question'], 'answers': [{'text': row['Answer']}]}]}, axis=1).tolist()

# 변환된 데이터를 JSON 파일로 저장
#/content/data_Inspection - 시트12 (1).csv
json_file_path = '/content/final-data.json'
with open(json_file_path, 'w', encoding='utf-8') as file:
    json.dump(qas_list, file, ensure_ascii=False)


# 저장된 JSON 파일 경로 반환
json_file_path
len(qas_list)
```

```python
import json

# JSON 파일을 읽습니다.
with open('/content/final-data.json', 'r') as f:
    dev_data = json.load(f)

# dev_data를 정제하여 refined_dict 생성
refined_dict = {}
for original_dict in dev_data:
    for entry in original_dict['qas']:
        question = entry['question']
        answers = [ans['text'] for ans in entry['answers']][0]
        refined_dict[question] = answers

print(refined_dict)
#여기서 문제생김. 왜?
```

```python
# 총 몇개의 쌍인지 확인
len(refined_dict)
```

```python
final_prompt_list = []

for idx, (question, answer) in enumerate(refined_dict.items()):
    prompt = f"Below is an instruction that describes a task. Write a response that appropriately completes the request. ### Instruction: {question} ### Response: {answer}"
    print(idx, prompt)
    prompt_dict = {}
    prompt_dict['text'] = prompt
    final_prompt_list.append(prompt_dict)
```

## Json 파일 형태로 저장 및 다운로드

```python
with open('/content/final-data.json', 'w', encoding='utf-8') as f:
    f.write("[\n")
    for i, item in enumerate(final_prompt_list):
        json_str = json.dumps(item, ensure_ascii=False)
        f.write(json_str)
        if i < len(final_prompt_list) - 1:  # 마지막 원소가 아니라면 콤마를 추가
            f.write(",\n")
        else:  # 마지막 원소라면 콤마를 추가하지 않음
            f.write("\n")
    f.write("]\n")
```
